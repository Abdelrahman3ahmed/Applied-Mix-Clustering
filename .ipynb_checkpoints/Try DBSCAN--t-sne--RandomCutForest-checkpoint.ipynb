```python
#function for t-sne Visualization
def T_sne_Visual(data , Num_Cluster):
    tsne = TSNE(n_components=2, verbose=0, perplexity=30, n_iter=300 , random_state=42)
    tsne_results = tsne.fit_transform(data)
    data['tsne-2d-one'] = tsne_results[:,0]
    data['tsne-2d-two'] = tsne_results[:,1]
    plt.figure(figsize=(16,10))
    sns.scatterplot(
        x="tsne-2d-one", y="tsne-2d-two",
        hue="y",
        palette=sns.color_palette("hls", Num_Cluster),
        data=data,
        legend="full",
        alpha=0.3
    )
```

```python
#Make label with Kmeans
def Make_labels_Clustering_tech(data1 , ep , mis_s):
    data = data1.copy()
    clustering = DBSCAN(eps= ep, min_samples=mis_s).fit(data)
    data["y"] = clustering.labels_
    n_clusters_ = len(set(clustering.labels_)) - (1 if -1 in clustering.labels_ else 0)
    print("DB nof Clusters " , n_clusters_)
    return data 
```

```python
def Expectation_Maximization(data1 , nof_Clus , qurat_prec):
    data = data1.copy()
    EM_model = mixture.GaussianMixture(n_components = nof_Clus)
    EM_model.fit(data)
#     data["y"] = EM_model.predict(data)
    
    threth = np.quantile(EM_model.score_samples(data) , qurat_prec)
    print("quantile" , threth)
    
    data["y"] = EM_model.score_samples(data)
    data.loc[(data.y > threth) , "y"] = 0
    data.loc[(data.y < threth) , "y"] = -1
    
    return data
```

```python
def Silhouette_score(data):
    d1_clustered = data.loc[(data.y >= 0)]
    print(silhouette_score(d1_clustered.iloc[:,:-1], d1_clustered["y"]))
    
```

```python
def Elow_method(data , mins):
    # we ask for mins + 1 nearest, because the data point it self (distance = 0) is included
    mins = mins
    nn = NearestNeighbors(n_neighbors= mins + 1)

    nn.fit(data)

    distances, neighbors = nn.kneighbors(data)

    distances
    # sort the distances
    distances = np.sort(distances[:, mins], axis = 0)
    
    #plot the distances
    distances_df = pd.DataFrame({"distances": distances,
                                 "index": list(range(0,len(distances)))})
    plt = (ggplot(distances_df, aes(x = "index", y = "distances")) +
     geom_line(color = "white", size = 2) + theme_minimal() +
     labs(title = "Elbow Method for Choosing eps") +
     theme(panel_grid_minor = element_blank(),
          rect = element_rect(fill = "#202124ff"),
          axis_text = element_text(color = "white"),
          axis_title = element_text(color = "white"),
          plot_title = element_text(color = "white"),
          panel_border = element_line(color = "darkgray"),
          plot_background = element_rect(fill = "#202124ff")
          ))
    ggsave(plot=plt, filename='elbow.png', dpi=300)

    return plt
     
```
